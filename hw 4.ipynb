```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as web
import seaborn as sns
```

```python
sp = pd.read_csv("../Phys 255/^GSPC.csv",header=0, delim_whitespace=True)

month_ave= 63
year_ave = 252
```

```python
#s_p500 = web.get_data_yahoo("^GSPC",start = "1970-10-10",end = "2020-10-10")


sp = pd.read_csv("../Phys 255/^GSPC1.csv")



sp = sp[['Date','Adj Close']]

sp.head()
```

```python
sp.rename(columns={'Date' :'date' ,'Adj Close' : 'price_t'}, inplace =True)
```

```python
sp.head()
```

```python
sp['price_t-1'] = sp['price_t'].shift(month_ave)
```

```python
sp.head()
```

```python
sp.set_index('date',inplace=True)
```

```python
sp['returns_manual'] = ((np.log(sp['price_t']) - np.log((sp['price_t-1']))/(sp['price_t-1']))) *100
```

```python
sp.head()
```

```python
sp['returns_manual'].plot(figsize=(12,8))
plt.legend('Q')
#quarterly
```

```python
sp['price_t-1'] = sp['price_t'].shift(year_ave)
sp['returns_manual'] = ((np.log(sp['price_t']) - np.log((sp['price_t-1']))/(sp['price_t-1']))) *100
sp['returns_manual'].plot(figsize=(12,8))
plt.legend('Annual')
#annual
```

```python
#s_p500 = web.get_data_yahoo("^GSPC",start = "1970-10-10",end = "2020-10-10")


sp = pd.read_csv("../Phys 255/^VIX1.csv")
sp = sp[['Date','Adj Close']]

sp.rename(columns={'Date' :'date' ,'Adj Close' : 'price_t'}, inplace =True)
sp['price_t-1'] = sp['price_t'].shift(month_ave)
sp['returns_manual'] = ((sp['price_t'] - (sp['price_t-1']))/(sp['price_t-1'])) *100
sp.set_index('date',inplace=True)
sp['returns_manual'].plot(figsize=(12,8))

sp['price_t-1'] = sp['price_t'].shift(year_ave)
sp['returns_manual'] = ((sp['price_t'] - (sp['price_t-1']))/(sp['price_t-1'])) *100

sp['returns_manual'].plot(figsize=(12,8))

```
